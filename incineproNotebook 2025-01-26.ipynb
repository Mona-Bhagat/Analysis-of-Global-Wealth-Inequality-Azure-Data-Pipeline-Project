```python

# Reading the Parquet file from the specified path into a dataFrame 'df', inferring schema including headers

df = spark.read.parquet("dbfs:/mnt/rawdata/dbo.economic-inequality-gini-index.parquet", header=True, inferSchema=True)
```

```python
## Dropping the column '_990179_annotations' from the DataFrame 'df' and creating a new DataFrame 'dropped_colum'

dropped_colum = df.drop('_990179_annotations')
```

```python
#displaying the contents of the DataFrame 'dropped colum' to see if the column was dropped (yes there is a speelling mistake ;-) )

display(dropped_colum)
```

```python
#Importing the 'round' function from the pyspark.sql.functions module for rounding numerical values
from pyspark.sql.functions import round
```

```python
## Rounding the 'Gini_coefficient' column to 2 decimal places in the DataFrame 'dropped_colum' and create a new DataFrame 'df_roundedgini'

df_roundedgini = dropped_colum.withColumn('Gini_coefficient', round('Gini_coefficient', 2))
```

```python
#displaying the contents of the DataFrame 'df_roundgini' to check if the rounding was successful
display(df_roundedgini)
```

```python
#Writing the 'df_roundedgini' dataFrame to a CSV file, overwriting any existing files at the specified location "/mnt/transformeddata"

df_roundedgini.write.mode("overwrite").csv("/mnt/transformeddata")
```
